# 9. Check number of unique values in each column, can they be combined in any way to ease encoding?  Comment your thoughts and make those changes.

```python
unique_values_counts = {}


for column in categorical_df.columns:
    unique_values = categorical_df[column].unique()
    value_counts = categorical_df[column].value_counts().to_dict()
    
    unique_values_counts[column] = {'unique_values': unique_values, 'value_counts': value_counts}

for column, info in unique_values_counts.items():
    print(f"Column: {column}")
    print("Unique Values:", info['unique_values'])
    print("Value Counts:", info['value_counts'])
```

1. Education: unify above master and doctor in "Master or above"

```python
values_to_unify = ['Master', 'Doctor']
unified_value = 'Master or above'
categorical_df['education'] = np.where(categorical_df['education'].isin(values_to_unify), unified_value, categorical_df['education'])
categorical_df["education"].unique()
```

2. vehicle_class: unify luxury

```python
categorical_df["vehicle_class"].unique()
```

```python
values_to_unify = ['Luxury SUV', 'Luxury Car']
unified_value = 'Luxury Car/SUV'
categorical_df["vehicle_class"] = np.where(categorical_df["vehicle_class"].isin(values_to_unify), unified_value, categorical_df["vehicle_class"])
categorical_df["vehicle_class"].unique()
```
